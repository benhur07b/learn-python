**Q**: Calculate precision and recall for the "Burned" class. Which is more important for this application?

**Definitions**:
- Precision = Percentage of predicted class that is correct (per class)
- Recall = Percentage of actual class correctly identified (per class)

<details>
<summary>Click to reveal answer</summary>

**Answers**:
- **Precision** = 90 / 150 = 0.60 (60%)
  - Of pixels predicted as burned, 60% actually were
- **Recall** = 90 / 100 = 0.90 (90%)
  - Of actual burned pixels, we found 90%

**More Important**: **Recall**

For fire damage assessment:
- **High recall** is critical: We don't want to miss burned areas (false negatives could delay aid)
- Lower precision is acceptable: False alarms can be verified with field checks
- It's better to overestimate damage than underestimate

## Coding Exercise 1: Crops VS Weed

**Scenario:** You have a small robot in a field. It takes a "Greenness Score" and a "Height" reading for every plant it sees. You want to train a model to distinguish between a Crop (what we want) and a Weed (what we want to remove).

### Step 1: Set Up the "Training Data"

Create a list of data points and a list of correct answers (labels).

- X (Features): Greenness and Height.
- y (Labels/Results): 0 for Weed, 1 for Crop.

Create a list for features: **[Greenness, Height]**

- Greenness = 80, 78, 25, 30
- Height = 10, 11, 2, 3

Create a list for labels:

- Average Crop Greenness: ~82
- Average Weed Greenness: ~32

<details>
<summary>Click to reveal code</summary>

```
# Each sub-list is [Greenness, Height]
# This is our 'Training Data'
training_data = [
    [80, 10], # Crop
    [78, 11], # Crop
    [25, 2],  # Weed
    [30, 3]   # Weed
]

# These are the correct answers (Labels)
labels = [1, 1, 0, 0]
```
</details>

### Step 2: Create the "Learning" Function

Write a function that acts as the "algorithm". It must look at the data and find the midpoint between the average Crop and the average Weed. This midpoint is the Decision Boundary.

<details>
<summary>Click to reveal code</summary>

```
def crop_weed_detect(data, labels):
    
    # Find all Greenness values for Crops (Label 1)
    crops = [data[i][0] for i in range(len(labels)) if labels[i] == 1]
    
    # Find all Greenness values for Weeds (Label 0)
    weeds = [data[i][0] for i in range(len(labels)) if labels[i] == 0]
    
    # Calculate the averages
    avg_crop = sum(crops) / len(crops)
    avg_weed = sum(weeds) / len(weeds)
    
    # Return the halfway point (The Pattern)
    return (avg_crop + avg_weed) / 2

# Train the model and save the pattern
learned_pattern = crop_weed_detect(training_data, labels)

```
</details>

### Step 3: Test on "Unseen" Data (Inference)

We have a new set of data with 4 plants. Create a script to use the "learned pattern" or "model" to label them.

Note: Use a for loop to process the new plants.

- **data** = [ [75, 9], [28, 4], [85, 12], [20, 1] ]


<details>
<summary>Click to reveal code</summary>

```
data = [[75, 9], [28, 4], [85, 12], [20, 1]]

for plant in data:
    greenness = plant[0]

    # Apply the learned_pattern or "model"
    if greenness > learned_pattern:
        print(f"Greenness {greenness}: Predicted as CROP")

    else:
        print(f"Greenness {greenness}: Predicted as WEED")

```
</details>
